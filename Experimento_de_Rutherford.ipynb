```python
#from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
def xt_n(xn_1,yn_1,vn_1,dt):
    m=1.67*10**-27
    k=9*10**9
    q=1.6*10**-19
    Q=79*q
    p=k*Q*q/m
    r=np.sqrt(xn_1**2 +yn_1**2)
    an_1=p*xn_1/r**3
    vn=vn_1 + an_1*dt
    return xn_1 +vn*dt+0.5*an_1*dt**2,vn

def yt_n(xn_1,yn_1,vn_1,dt):
    m=1.67*10**-27
    k=9*10**9
    q=1.6*10**-19
    Q=79*q
    p=k*Q*q/m
    r=np.sqrt(xn_1**2 +yn_1**2)
    an_1=p*yn_1/r**3
    vn=vn_1 + an_1*dt
    return yn_1 +vn*dt + 0.5*an_1*dt**2,vn
```

```python
m=1.67*10**-27
K=4.7*10**6*1.6*10**-19

yns=np.linspace(-10**-13,10**-13,100)
xs=[]
ys=[]
thethas=[]
for yn in yns:
    xn=-2000*10**-15
    vx0=np.sqrt(2*K/m)
    vy0=0
    dt=10**-21
    x=[]
    y=[]
    for _ in range(1000):
        xn,vx0=xt_n(xn,yn,vx0,dt)
        yn,vy0=yt_n(xn,yn,vy0,dt)
        x.append(xn)
        y.append(yn)
    thethas.append(np.pi-np.arctan(y[-1]/x[-1]))
    xs.append(x)
    ys.append(y)


plt.figure(figsize=(30,15))
for x,y in zip(xs,ys):
    plt.plot(x,y,lw=3,color="red")
plt.title("Trayectorias de la particula positiva para diferentes valores de b",fontsize=20)
plt.scatter(0, 0, s=500, color="blue")
plt.axhline()
plt.axvline()
```

```python
plt.figure(figsize=(16,8))
plt.scatter(yns,thethas)
plt.axvline()
plt.xlabel("b",fontsize=20)
plt.ylabel("thetha",fontsize=20)
```
